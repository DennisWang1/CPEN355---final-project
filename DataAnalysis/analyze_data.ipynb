```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
image_table = pd.read_csv('D:\\CPEN355_project\\Data\\Image_table.csv')
genmodel_table = pd.read_csv('D:\\CPEN355_project\\Data\\Price_table.csv')

# Analyze the distribution of `Genmodel_ID` in image_table
genmodel_counts = image_table['Genmodel_ID'].value_counts()

# Plot the distribution of the bottom 20 `Genmodel_IDs`
bottom_20 = genmodel_counts.tail(20)

# Print the least 20 Genmodel_IDs and their sample counts
print("Least 20 Genmodel_IDs and their counts:")
print(bottom_20)

# Plot bar chart
plt.figure(figsize=(10, 6))
bottom_20.plot(kind='bar')
plt.title('Bottom 20 Genmodel_IDs by Sample Count')
plt.xlabel('Genmodel_ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analyze the distribution of years in the vehicle model table
year_counts = genmodel_table['Year'].value_counts().sort_index()

# Plot the year distribution bar chart
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar')
plt.title('Distribution of Years in Genmodel Table')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Check the distribution of the quality check field
quality_counts = image_table['Quality_check'].value_counts()

# Plot the distribution of quality check results
plt.figure(figsize=(10, 6))
quality_counts.plot(kind='bar')
plt.title('Distribution of Quality Check in Image Table')
plt.xlabel('Quality Check')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print simple statistics for distribution analysis
print("Genmodel_ID Distribution:")
print(genmodel_counts)
print("\nQuality Check Distribution:")
print(quality_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
image_table = pd.read_csv('D:\\CPEN355_project\\Data\\Image_table.csv')

# Analyze the distribution of `Genmodel_ID` in the image_table
genmodel_counts = image_table['Genmodel_ID'].value_counts()

# Count the number of Genmodel_IDs in different intervals of image counts
bins = [0, 10, 20, 30, 40, 50, 100, 200, 500, 1000, genmodel_counts.max() + 1]
labels = ['1-10', '11-20', '21-30', '31-40', '41-50', '51-100', '101-200', '201-500', '501-1000', '1000+']
genmodel_bins = pd.cut(genmodel_counts, bins=bins, labels=labels, right=False)
genmodel_bin_counts = genmodel_bins.value_counts(sort=False)

# Print statistics
print("Total number of Genmodel_IDs:", len(genmodel_counts))
print("\nGenmodel_ID count by image number intervals:")
print(genmodel_bin_counts)

# Plot the distribution of Genmodel_IDs by image count intervals
plt.figure(figsize=(10, 6))
genmodel_bin_counts.plot(kind='bar')
plt.title('Distribution of Genmodel_IDs by Image Count Intervals')
plt.xlabel('Number of Images Interval')
plt.ylabel('Number of Genmodel_IDs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
